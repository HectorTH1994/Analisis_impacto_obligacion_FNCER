```python
import requests

# URL de la API de Nominatim
url = "https://nominatim.openstreetmap.org/search"

# Parámetros de búsqueda para todas las fábricas de acero en Colombia
params = {
    "q": "fabrica de acero",
    "countrycodes": "co",
    "format": "json",
    "limit": 1000 # Límite máximo de resultados
}

# Realizar la solicitud HTTP a la API de Nominatim
response = requests.get(url, params=params)

print(response.json())

# Obtener las coordenadas de los resultados de la búsqueda
coordinates = [(float(result["lat"]), float(result["lon"])) for result in response.json()]

# Imprimir las coordenadas obtenidas
print(coordinates)
```

```python
import requests
import pandas as pd

# Definimos un cuadrado que contenga nuestra zona de interes
lon_min = -75.318
lat_min = 7.538
lon_max = -72.117
lat_max = 11.517

# URL de la API de Nominatim
url = "https://nominatim.openstreetmap.org/search"

# Parámetros de búsqueda para todas las fábricas de acero en Colombia
params = {
    "q": "Ternium",
    "format": "json",
    "limit": 1000, # Límite máximo de resultados
    "accept-language": "es", # Configurar la respuesta en español
   "viewbox": "-74.7677,10.375,-73.4425,11.4938" # Limitar la búsqueda a las regiones deseadas
}

# Realizar la solicitud HTTP a la API de Nominatim
response = requests.get(url, params=params)

# Obtener los nombres y coordenadas de los resultados de la búsqueda
results = []
for result in response.json():
    name = result["display_name"]
    lat = float(result["lat"])
    lon = float(result["lon"])
    results.append({"name": name, "lat": lat, "lon": lon})

# Convertir los resultados a un DataFrame de pandas
df = pd.DataFrame(results)

# Imprimir el DataFrame
print(df)
```

```python
import requests
import pandas as pd
import folium

# Definir la URL de la API de Nominatim y los parámetros de búsqueda
url = "https://nominatim.openstreetmap.org/search"
params = {
    "q": "poliductos y gaseoductos en Colombia",
    "format": "json",
    "limit": 1000,
    "accept-language": "es",
}

# Realizar la solicitud HTTP a la API de Nominatim y obtener las coordenadas de los resultados
response = requests.get(url, params=params)
results = response.json()
coordinates = [(float(result["lat"]), float(result["lon"])) for result in results]

# Crear un objeto de mapa de Folium centrado en Colombia
colombia_map = folium.Map(location=[4.5709, -74.2973], zoom_start=6)

# Agregar un marcador para cada par de coordenadas obtenido
for coord in coordinates:
    folium.Marker(location=coord).add_to(colombia_map)

# Mostrar el mapa
colombia_map

```
